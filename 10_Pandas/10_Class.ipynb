### Filling the value at NAN rows or position

```python
df = iris.copy()
df.columns = ['sl','sw','pl','pw','ft']
df.head()
```

```python
df.iloc[2:4,1:3] = np.nan
df.head()
```

```python
df.dropna(inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.iloc[2:4,1:3] = np.nan
df.head()
```

```python
df.sw.fillna(df.sw.mean(), inplace=True)
# df.fillna({'pl':df.pl.mean()}, inplace=True)
df.head()
```

```python
df.fillna({'sw':df.sw.mean()}, inplace=True)
df.fillna({'pl':df.pl.mean()}, inplace=True)
df.head()
```

```python
df.iloc[2:4,1:3] = np.nan
df.head()
```

```python
sw_df = df.ftype == 'Setosa'
print(df[sw_df].head())

df.fillna({'sw':df[sw_df].sw.mean()}, inplace=True)
df.head()
```

```python
pl_df = df.ftype == 'Setosa'
print(df[pl_df].head())

df.fillna({'pl':df[pl_df].pl.mean()}, inplace=True)
df.head()
```
