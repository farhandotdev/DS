```python
x=np.arange(0,5,0.1)
y1=x**2
y2=x**3

plt.plot(x,y1, color='red', linewidth=3, label="y1 = x^2")
plt.plot(x,y2, color='green', linewidth=3, label="y2 = x^3")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Matplot Demo')
plt.legend()
plt.grid()
plt.show()
```

#### Bubble Chart

```python
import matplotlib.pyplot as plt
import numpy as np

years = [2012,2013,2014,2015,2016,2017]
salary = [12,14,13,15,17,20]
population = [100,200,300,400,420,480]

colors = np.arange(len(years))

# plt.scatter(years, salary, s=population, alpha=0.5, edgecolors='r', marker='^')
plt.scatter(years, salary, s=population, c=colors, alpha=0.5, edgecolors='black', )
plt.text(years[0] + 0.2, salary[0], population[0])
plt.text(years[1] + 0.2, salary[1], population[1])
plt.text(years[2] + 0.2, salary[2], population[2])
plt.text(years[3] + 0.2, salary[3], population[3])
plt.text(years[4] + 0.2, salary[4], population[4])
plt.text(years[5] + 0.2, salary[5], population[5])
plt.title('Bubble Chart')
plt.show()
```

#### Pie Chart

```python
sizes = [2,3,5,6]
labels = ['A','B','C','D']
colors = ['yellow','pink','gray','green']
explode = [0.1,0.2,0,0]
plt.title('Pie Chart')
plt.pie(sizes, colors=colors, labels=labels, counterclock=False, explode=explode, autopct='%.2f%%', startangle=90)
plt.show()
```
