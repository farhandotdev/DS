Find value of killed column only where country == 'United State'? and
Find total number of people killed from USA?

```python
import numpy as np
import csv

with open('terrorismData.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)

    killed = []
    country = []
    for i in file_list:
        killed.append(i['Killed'])
        country.append(i['Country'])

    np_killed = np.array(killed)
    np_killed[np_killed == ''] = '0.0'

    np_killed = np.array(np_killed, dtype=float)

    np_country = np.array(country)
    bool_arr = np_country == 'United States'

    for i in np_killed[bool_arr]:
        print(int(i))

    print(int(sum(np_killed[bool_arr])))
```

Find the number of attack held between day 10 and day 20?

```python
import csv

with open('terrorismData.csv') as file_obj:
    file_data = csv.reader(file_obj, skipinitialspace=True)
    file_list = list(file_data)
    killed = []
    for i in file_list[1:]:
        day = int(float(i[2]))
        if day>=10 and day<=20:
            killed.append(1)
        
print(sum(killed))
        
```

Find the number of attack held between 1 Jan 2010 and 31 Jan 2010?(including both date).

```python
import csv

with open('terrorismData.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)

    year = []

    for i in file_list[1:]:
        if i['Year'] == '2010' and i['Month'] == '1' and i['Day'] != '0':
            year.append(1)
    
    print(sum(year))
```

Terror Casualty attack in 'Jammu and Kashmir'

```python
import csv

with open('terrorismData.csv', encoding='utf-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)

    casualty= []
    city = []
    terroristGroup = []

    for row in file_data:
        if row['Year'] == '1999' and row['State'] == 'Jammu and Kashmir':
            if row['Month'] >= '5' and row['Month'] <= '7':
                if row['Killed'] == '' :
                    row['Killed'] = '0'   
                if row['Wounded'] == '' :
                    row['Wounded'] = '0'
                if row['City'] !='Unknown' and row['Group'] != 'Unknown' :
                        casualty.append(int(float(row['Killed'])) + int(float(row['Wounded'])))
                        city.append(row['City'])
                        terroristGroup.append(row['Group'])

ind = casualty.index(max(casualty))
print(casualty[ind], city[ind], terroristGroup[ind])
```

Terror casualty red states include Jharkhand, Odisha, Andhra Pradesh, and Chhattisgarh?

```python
import numpy as np
import csv

with open('terrorismData.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)

    killed = []
    wounded = []

    for i in file_list:
        if i['State'] == 'Jharkhand' or i['State'] == 'Odisha' or i['State'] == 'Andhra Pradesh' or i['State'] == 'Chhattisgarh':    
            killed.append(i['Killed'])
            wounded.append(i['Wounded'])

    np_killed = np.array(killed)
    np_killed[np_killed == ''] = '0'

    np_wounded = np.array(wounded)
    np_wounded[np_wounded == ''] ='0'

    np_killed = np.array(np_killed, dtype=float)
    np_wounded = np.array(np_wounded, dtype=float)

    casualty = np_killed + np_wounded

print(int(sum(casualty)))
```

Find top 5 indian cities which has most number of casualties?

```python
import numpy as np
import csv
from collections import Counter

with open('terrorismData.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)

    city = []
    casualty = []

    for i in file_list:
        if i['City'] != 'Unknown' and i['Country'] == 'India':
            casualty.append([i['Killed'], i['Wounded']])
            city.append(i['City'])
        
    np_city = np.array(city)
    np_casualty = np.array(casualty)

    np_casualty[np_casualty == ''] = '0'
    np_casualty = np.array(np_casualty, dtype=float)
    np_casualty = np.sum(np_casualty, axis=1)
    
dic = {}
for i in range(len(np_city)):
    if np_city[i] in dic:
        dic[np_city[i]] = dic[np_city[i]] + np_casualty[i]
    else:
        dic[np_city[i]] = np_casualty[i]
        
k = Counter(dic)
high = k.most_common(5)

for i in high:
    print(i[0], int(i[1]))

```

Find the most frequent day of attack in a terrorismDataset?

```python
import numpy as np
import csv

with open('terrorismData.csv') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    day = []

    for i in file_data:
        day.append(i['Day'])
        
    np_day = np.array(day, dtype=int)
    day, count = np.unique(np_day, return_counts=True)
    ind = np.argmax(count)
    print(day[ind], count[ind])
```
